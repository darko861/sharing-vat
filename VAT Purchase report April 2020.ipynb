```python
def vat_amount(row):
    
    vat = row[39]
    
    if vat >0:
        return 'VAT'
    elif vat <0:
        return 'VAT'
    else:
        return 'Transactions 0DKK'

vat['Comments'] = vat.apply(vat_amount, axis='columns')
```
