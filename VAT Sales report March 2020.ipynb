```python
def vat_amount(row):

    vat_yes = row[34]

    if vat_yes > 0:
        return "VAT"
    elif vat_yes < 0 :
        return "VAT"
    else:
        return "Transactions 0DKK"
    
vat["Comment"] = vat.apply(vat_amount, axis= "columns")    
```

```python
#pd.options.display.max_rows = 4000
vat.head(20)
```

```python
def applyFunc(s):
    if s == "VAT":
        return "Domestic 25%"
    

vat["Transaction type"] = vat["Comment"].apply(applyFunc)
```
